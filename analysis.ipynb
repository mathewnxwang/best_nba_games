### Data Transformation

```python
# select and filter to relevant data

game_df = data['game'][['game_id', 'game_date', 'team_name_home', 'team_name_away', 'season_type']]
game_df = game_df[game_df['season_type'].isin(['Regular Season', 'Playoffs'])]
game_df = game_df[game_df['game_date'] >= '2017-01-01']

line_score_df = data['line_score']\
    [['game_id', 'team_wins_losses_home', 'team_wins_losses_away', 'pts_ot1_home', 'pts_ot2_home', 'pts_ot1_away', 'pts_ot2_away',
      'pts_home', 'pts_away']]

other_stats_df = data['other_stats'][['game_id', 'lead_changes', 'times_tied']]
```

```python
# join data together
merged_df = pd.merge(game_df, line_score_df, on='game_id')
merged_df = pd.merge(merged_df, other_stats_df, on='game_id')
```

### Feature Engineering

```python
def record_games_n(record):
    record_list = record.split('-')
    record_int = [int(e) for e in record_list]
    game_n = sum(record_int)
    return game_n
```

```python
# create feature for regular season win percentages

def compute_win_percentage(record):
    if record == '-':
        return 0
    record_list = record.split('-')
    wins, losses = map(int, record_list)
    win_perc = wins / (wins + losses)
    return win_perc

# find win percentage only for regular season games

regular_season_games = merged_df.loc[merged_df['season_type'] == 'Regular Season']
merged_df.loc[regular_season_games.index, 'reg_season_win_pct_home'] = regular_season_games['team_wins_losses_home'].apply(compute_win_percentage)
merged_df.loc[regular_season_games.index, 'reg_season_win_pct_away'] = regular_season_games['team_wins_losses_away'].apply(compute_win_percentage)
```

```python
playoff_games = merged_df.loc[merged_df['season_type'] == 'Playoffs']

# create feature for playoff series game #
merged_df.loc[playoff_games.index, 'series_game_n'] = playoff_games['team_wins_losses_home'].apply(record_games_n)

# create feature for record diff for playoff series
def compute_win_diff(record):
    record_list = record.split('-')
    team1_wins, team2_wins = map(int, record_list)
    diff = abs(team1_wins - team2_wins)
    return diff

merged_df.loc[playoff_games.index, 'playoff_win_diff'] = playoff_games['team_wins_losses_home'].apply(compute_win_diff)
```

```python
# create feature for overtime period

def define_ot_period(row):
    ot2_pts = row['pts_ot2_home'] + row['pts_ot2_away']
    if ot2_pts > 0:
        return 'OT2+'
    ot1_pts = row['pts_ot1_home'] + row['pts_ot1_away']
    if ot1_pts > 0:
        return 'OT1'
    return 'Reg'

merged_df['ot_period'] = merged_df.apply(define_ot_period, axis=1)
```

```python
# create feature for how close games ended up

def define_close_games(row):
    points_diff = abs(row['pts_home'] - row['pts_away'])
    return points_diff

merged_df['end_points_diff'] = merged_df.apply(define_close_games, axis=1)
```

### Normalization

Create a score for each feature

```python
# regular season win percentages
def normalize_win_percentage(row):

    # don't score games very early on in the season, since records are too noisy
    games_n = record_games_n(row['team_wins_losses_home'])
    too_few_games = games_n <= 15

    if not too_few_games and row['reg_season_win_pct_home'] >= 0.75 and row['reg_season_win_pct_away'] >= 0.75:
        return 2
    if not too_few_games and row['reg_season_win_pct_home'] >= 0.6 and row['reg_season_win_pct_away'] >= 0.6:
        return 1
    return 0

merged_df['reg_season_win_pct_score'] = merged_df.apply(normalize_win_percentage, axis=1)
```

```python
# playoff series game #
def normalize_series_game_n(game_n):
    if game_n == 7:
        return 1.5
    if game_n >= 5:
        return 0.75
    return 0

merged_df['playoffs_game_n_score'] = merged_df['series_game_n'].apply(normalize_series_game_n)
```

```python
# OT period
def normalize_ot_period(ot_period):
    if ot_period == 'OT2+':
        return 1
    if ot_period == 'OT1':
        return 0.5
    return 0

merged_df['ot_period_score'] = merged_df['ot_period'].apply(normalize_ot_period)
```

```python
# Close games
def normalize_close_games(pts_diff):
    if pts_diff <= 3:
        return 1
    if pts_diff <= 6:
        return 0.5
    return 0

merged_df['close_game_score'] = merged_df['end_points_diff'].apply(normalize_close_games)
```

```python
# Lead changes
def normalize_lead_changes(lead_changes):
    if lead_changes > 20:
        return 0.5
    if lead_changes > 10:
        return 0.25
    return 0

merged_df['lead_change_score'] = merged_df['lead_changes'].apply(normalize_lead_changes)
```

### Model Scoring

```python
def calculate_score(row):

    score = row['ot_period_score'] + row['close_game_score'] + row['lead_change_score']

    # calculate score for regular season games
    if row['season_type'] == 'Regular Season':
        score += row['reg_season_win_pct_score']
        return score
    
    score += row['playoffs_game_n_score']
    return score

merged_df['model_score'] = merged_df.apply(calculate_score, axis=1)
```
